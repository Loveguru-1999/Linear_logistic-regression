```python
import numpy as np
import pandas as pd
import seaborn as sns
import pandas_profiling
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
xx=pd.read_csv('USA_Housing.csv')
```

```python
xx.head(3)
```

```python
xx.columns
```

```python
xx.columns
```

```python
x=xx[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
y=xx['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
l=LinearRegression()
```

```python
l.fit(x_train,y_train)
```

```python
l.intercept_
```

```python
l.coef_
```

```python
predi=l.predict(x_test)
```

```python
predi
```

```python
plt.scatter(y_test,predi)
```

```python
sns.distplot(predi-y_test)
```

```python
from sklearn import metrics
```

```python
metrics.mean_squared_error(y_test,predi)
```

```python
metrics.mean_squared_log_error(y_test,predi)
```
